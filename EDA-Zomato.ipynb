```python
final=pd.merge(data,data_country,on='Country Code',how='left')
```

```python
final.head()
```

```python
final.columns
```

```python
country_names=final.Country.value_counts().index
```

```python
final.Country.value_counts().index
```

```python
x=final.Country.value_counts().values
```

```python
plt.pie(x[:3],labels=country_names[:3],autopct='%1.2f%%')
```

### Observations:
Maximum number of users are from india and then USA and UK(94.39%, 4.73%, 0.87%)

```python
final.columns
```

```python
rating_df=final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
rating_df
```

## Observation
    Rating of 0 means not rated
    Rating between 0 and 2.4 means Poorly rated
    Rating between 2.5 and 3.4 means average rated
    Rating between 3.5 and 3.9 means good rated
    Rating between 4.0 and 4.4 means Very good rated
    Rating between 4.5 and 4.9 means Excellent Rated
    


```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12, 6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=rating_df)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=rating_df,palette=['white','red','orange','yellow','green','green'])
```

```python
sns.countplot(x='Rating color',data=rating_df,palette=['white','red','orange','yellow','green','green'])
```

## Observation
    Many People haven't rated
    Maximung rated as 2.5 to 3.4
